Flight info with Seaborn

```python
sns.get_dataset_names()
flights = sns.load_dataset('flights')
flights
```

```python
sns.relplot(x='year', y='passengers', hue='month', data=flights)
```

```python
year_sums = flights.groupby('year').sum()
sns.relplot(x='year', y='passengers', data=year_sums)
```

```python
sns.lmplot(x='year', y='passengers', data=flights)
print()
```

```python
from calendar import month_abbr
g = sns.barplot(x='month', y='passengers', data=flights)
g.set_xticklabels(labels=list(month_abbr)[1:], rotation=45)
print()
```

```python
anscombe = sns.load_dataset("anscombe")
sns.regplot(x='x', y='y', data=anscombe.query("dataset == 'I'"))
```
